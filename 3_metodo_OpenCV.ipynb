```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
## função para subtrair as imagens com absdiff e ver a diferença de pixels das imagens
def compare_images (image1_path, image2_path, image3_path):
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)
    image3 = cv2.imread(image3_path)

    ## Verifica se as imagens foram carregadas
    if image1 is None or image2 is None or image3 is None:
        print("Erro ao carregar as imagens")
        return

    ## Converte para escala de cinza
    gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
    gray_image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)

    ## Subtrai os pixel entre duas imagens (ex: pixel_imagem1 - pixel_imagem2)
    diff_image1_2 = cv2.absdiff(gray_image1, gray_image2)
    diff_image2_3 = cv2.absdiff(gray_image2, gray_image3)
    diff_image1_3 = cv2.absdiff(gray_image1, gray_image3)

    ## Se a diferença entre os pixels de duas imagens for menor que 30, elas serão consideradas semelhantes
    threshold = 30
    similar1_2 = (diff_image1_2 < threshold).all()
    similar2_3 = (diff_image2_3 < threshold).all()
    similar1_3 = (diff_image1_3 < threshold).all()

    ## Configuração do layout
    fig, axs = plt.subplots(2, 3, figsize=(30, 8))
    fig.suptitle("Comparação de Imagens", fontsize=30)

    ## Organização do layout
    axs[0, 0].imshow(gray_image1, cmap='gray')
    axs[0, 0].set_title('Imagem 1')
    axs[0, 0].axis('off')

    axs[0, 1].imshow(gray_image2, cmap='gray')
    axs[0, 1].set_title('Imagem 2')
    axs[0, 1].axis('off')

    axs[0, 2].imshow(gray_image3, cmap='gray')
    axs[0, 2].set_title('Imagem 3')
    axs[0, 2].axis('off')

    axs[1, 0].imshow(diff_image1_2, cmap='gray')
    axs[1, 0].set_title('Diferença 1-2')
    axs[1, 0].axis('off')

    axs[1, 1].imshow(diff_image2_3, cmap='gray')
    axs[1, 1].set_title('Diferença 2-3')
    axs[1, 1].axis('off')

    axs[1, 2].imshow(diff_image1_3, cmap='gray')
    axs[1, 2].set_title('Diferença 3-1')
    axs[1, 2].axis('off')

    ## ajustar automaticamente o espaçamento
    plt.tight_layout()
    plt.show()
    print()

    ## Verifica se as imagens são consideradas semelhantes ou diferentes
    if similar1_2:
        print("A imagem 1 é semelhante à imagem 2;")
    else:
        print("A imagem 1 é diferente da imagem 2;")

    if similar2_3:
        print("A imagem 2 é semelhante à imagem 3;")
    else:
        print("A imagem 2 é diferente da imagem 3;")

    if similar1_3:
        print("A imagem 3 é semelhante à imagem 1.")
    else:
        print("A imagem 3 é diferente da imagem 1.")
```
